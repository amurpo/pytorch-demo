```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import time

# Configurar el grfico en tiempo real
plt.ion()
fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot(x_data, y_data)

# Generar datos en tiempo real
for i in range(100):
    # Generar datos de ejemplo con PyTorch
    x = torch.tensor(i, dtype=torch.float32)
    y = torch.sin(x)
    
    # Actualizar los datos del grfico
    x_data.append(x.item())
    y_data.append(y.item())
    line.set_xdata(x_data)
    line.set_ydata(y_data)
    
    # Actualizar lmites del grfico
    ax.relim()
    ax.autoscale_view()
    
    # Redibujar el grfico
    fig.canvas.draw()
    fig.canvas.flush_events()
    
    # Pausa para que sea visible en tiempo real
    time.sleep(0.1)

# Mantener el grfico abierto al finalizar
plt.ioff()
plt.show()
```
