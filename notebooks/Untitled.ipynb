```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import time

# Configurar el gráfico en tiempo real
plt.ion()
fig, ax = plt.subplots()
x_data, y_data1, y_data2 = [], [], []
line1, = ax.plot(x_data, y_data1, label='Sin(x) * exp(-x/50)')
line2, = ax.plot(x_data, y_data2, label='Cos(x) * exp(-x/50)')

# Ajustar detalles del gráfico
ax.set_title('Real-time Plot of Complex Functions')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend()

# Generar datos en tiempo real
for i in range(100):
    # Generar datos de ejemplo con PyTorch
    x = torch.tensor(i, dtype=torch.float32)
    y1 = torch.sin(x) * torch.exp(-x / 50)
    y2 = torch.cos(x) * torch.exp(-x / 50)
    
    # Actualizar los datos del gráfico
    x_data.append(x.item())
    y_data1.append(y1.item())
    y_data2.append(y2.item())
    line1.set_xdata(x_data)
    line1.set_ydata(y_data1)
    line2.set_xdata(x_data)
    line2.set_ydata(y_data2)
    
    # Actualizar límites del gráfico
    ax.relim()
    ax.autoscale_view()
    
    # Redibujar el gráfico
    fig.canvas.draw()
    fig.canvas.flush_events()
    
    # Pausa para que sea visible en tiempo real
    time.sleep(0.1)

# Mantener el gráfico abierto al finalizar
plt.ioff()
plt.show()

```
